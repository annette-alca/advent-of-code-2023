```python
#Day25
import pandas as pd
```

```python
#do example first
file = open('example25.txt','r')
exlines = file.read().splitlines()
file.close()
# for i in range(len(exlines)):
#     l = exlines[i].replace('\n','')
#     exlines[i] = l
exlines
```

```python
#get all unique components (3 letter codes)
comps = []
compcount = {} #times a connection is made
for line in exlines:
    clist = line.split(' ')
    for c in clist:        
        c1 = c.replace(':','')
        if c1 not in comps:
            comps.append(c1)
            compcount[c1] = 1
        else:
            compcount[c1] +=1
compcount
```

```python
#table of connections
connectdf = pd.DataFrame(columns=comps)
connectdf.insert(0,'crosscomp',comps)
connectdf.fillna('',inplace=True)
connectdf
```

```python
for line in exlines:
    connect = line.split(': ')
    col = connect[0]
    rowlist = connect[1].split(' ')
    for r in rowlist:
        connectdf.loc[comps.index(r),col] = True
        connectdf.loc[comps.index(col),r] = True
connectdf
    
```
